# Write a Python program to removes multiple items from a given set

## Example
my_set: {1, 2, 3, 4, 5}

items to be removed: {1, 4}

Output: {2, 3, 5}

```python
my_set = {1,2,3,4,5}
remove = {1,4}
my_set = my_set - remove
print(my_set)
```

# find the factorial of a number 
## Example
(factorial of 5) = 5 * 4 * 3 * 2 * 1 = 120

```python
number = int(input("Enter a number: "))
factorial = 1
for i in range(1, number + 1):
    factorial *= i
print("(Factorial of)", number, "=", factorial)
```

# Create a function that Calculates the Sum of Even Numbers in a List
## Example
numbers: [1, 2, 3, 4, 5, 6]

Output: 12

```python
numbers = [1, 2, 3, 4, 5, 6]
def sum_OF_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total
result = sum_OF_numbers(numbers)
print(result)
```

# Create a function that takes a list as input and return reversed list

# create a function that calculates the median of a list of numbers (a median is the middle element in sorted list of numbers)

## Example
numbers: [5, 2, 7, 3, 1]

Output: 3

```python
numbers = [5, 2, 7, 3, 1]
def median(numbers):
    numbers.sort()  ## the list after sorting -->> numbers = [1, 2, 3, 5, 7]
    new_num = len(numbers) # len function i used it to calculate the number of elements in the list
    mid = new_num // 2
    return numbers[mid]
print(median(numbers))
```
