```python
#########----TASK_1----#########
email = "Amit_ml@gmail.edu"
parts = email.split("@") #-->["Amit_ml", "gmail.edu"]
username = parts[0]      #--->Amit_ml
domain_parts = parts[1].split(".")   #-----["gmail", "edu"]
domain = domain_parts[0]    #gmail
extension = domain_parts[-1]    #edu
if extension == "com":
    domain_type = "Commercial Domain"
elif extension == "edu":
    domain_type =  "Educational Domain"
else:
    domain_type = "Other Domain"
print("Username is:", username)
print("Domain is:", domain)
print("Domain Type is:", domain_type)
```

```python
#########----TASK_2----#########
msg = "###!!@mocleW EPGTQ!!6789"
clean = msg[6:18] #M - ! ----->mocleW EPGTQ
first_word = clean[:6]     # start from index 0 to index 6 -->moclew
second_word = clean[7:]    # start from index-7 to end of clean
first_word = first_word[::-1] # reverse
print(first_word + " " + second_word)
```

```python
##########----TASK_3----#########
msg = "&&&**$gnirtS PLIO!@1234"
core = msg[6:17]          # gnirts PLIO
first_word = core[:6]     # gnirts
second_word = core[7:]    # PLIO
first_word = first_word[::-1]   # reverse
second_word = second_word.replace("I", "E")
second_word = second_word.replace("O", "U")
print(first_word + " " + second_word)
```
